```python
import glob
import os
```

```python
data_path = "/content/dataset/training_set"
```

```python
cats_path = os.path.join(data_path, "cats/")
dogs_path = os.path.join(data_path, "dogs/")
```

```python
os.listdir(cats_path)
```

```python
cats_files = glob.glob(cats_path + "*.jpg")
dogs_files = glob.glob(dogs_path + "*.jpg")
```

```python
cat_pic = cv2.imread(cats_files[15])
```

```python
cat_pic.shape
```

```python
plt.imshow(cat_pic)
```

```python
cats_img, dogs_img = [], []

for file in cats_files:
  img = cv2.imread(file)
  img = cv2.resize(img, (224, 224))
  cats_img.append(img)

for file in dogs_files:
  img = cv2.imread(file)
  img = cv2.resize(img, (224, 224))
  dogs_img.append(img)
```

```python
len(cats_img), len(dogs_img)
```

```python
cats_labels = [0 for i in range(len(cats_files))]
dogs_labels = [1 for i in range(len(dogs_files))]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
X = cats_img + dogs_img
y = cats_labels + dogs_labels
y = to_categorical(y)

X, y = np.asarray(X), np.asarray(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.applications import resnet50
```

```python
base_model = resnet50.ResNet50(include_top=False, weights="imagenet", pooling="avg")
```

```python
model = Sequential()
```

```python
model.add(base_model)
model.add(Dense(2, activation="softmax"))
```

```python
for layer in model.layers[:-1]:
  layer.trainable = False
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, validation_split=0.15, batch_size=24, epochs=10)
```
